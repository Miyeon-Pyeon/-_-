## 3회 기출

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p3_1.csv')
df.head()
```

```python
# 1
# A등급에 해당하는 유저는 몇명인지 확인하라

df['dels'] = df['투약후'] - df['투약전']

def groupby_filter(x):
    if x <=-3:
        return 'A'

    elif x <=0:
        return 'B'

    else:
        return 'C'

df['groupby'] = df['dels'].apply(groupby_filter)
result = df['groupby'].value_counts().sort_index()['A']
print(result)
```

```python
target = df['groupby'].value_counts().sort_index().to_frame()
target['expected'] = [target['groupby'].sum()*0.5,target['groupby'].sum()*0.25,target['groupby'].sum()*0.25]

from scipy.stats import chisquare
s,p = chisquare(target['groupby'],target['expected'])

round_s = round(s,3)
print(round_s)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e3_p3_2_.csv')
df.head()
```

```python
# 1
# A,B 공장 각각 정규성을 가지는지 샤피로 검정을 통해 확인하라. (각 공장의 pvalue 출력할 것)

from scipy.stats import shapiro

a = df[df['group'] == 'A'].rpm
b = df[df['group'] == 'B'].rpm

print(shapiro(a))
print(shapiro(b))
```

```python
# 2
# A,B 공장 생산 제품의 rpm은 각각 정규성을 가지는지 샤피로 검정을 통해 확인하라. (각 공장의 pvalue 출력할 것)

from scipy.stats import levene

s, p = levene(a,b)

print(round(p, 3))
```

```python
# 3
# 대응 표본 t 검정을 통해 B공장 제품들의 rpm이 A 공장 제품의 rpm보다 크다고 말할 수 있는지 검정하라. pvalue를 소숫점 이하 3자리까지 출력하고 귀무가설, 대립가설 중 하나를  출력하라

from scipy.stats import ttest_rel
s, p = ttest_rel(b, a, alternative='greater')
print(round(p,3))
```

## 2회 기출

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e2_p3_1.csv')
df.head(5)
```

```python
print(df.info())
```

```python
df['height']  = df['height'].astype('str')
df
```

```python
# 1
# 122마리의 height 평균값을 m(미터) 단위로 소숫점 이하 5자리까지 실수 값만  출력하라

df['height_raw'] =df['height'].str.replace('cm','').astype('float')
result = round(df['height_raw'].mean()/100,5)
result
```

```python
# 2
# 모집단의 평균 길이가 30cm 인지 확인하려 일표본 t 검정을 시행하여 확인하려한다. 검정통계량을 소숫점 이하 3째자리까지 구하여라

from scipy.stats import ttest_1samp

s, p = ttest_1samp(df['height_raw'], 30)
print(round(s, 3))
```

```python
# 3
# 위의 통계량에 대한 p-값을 구하고 (반올림하여 소숫점 이하 3째자리), 유의수준 0.05하에서 귀무가설과 대립가설중 유의한 가설을 하나를 선택하시오(귀무/대립)

print(round(p, 3))

# 0.05보다 크므로 귀무가설이 유의
```

```python
# 조사결과 70%의 성인 남성이 3년 동안에 적어도 1번 치과를 찾는다고 할때, 21명의 성인 남성이 임의로 추출되었다고 하자.
```

```python
# 4
# 21명 중 16명 미만이 치과를 찾았을 확률(반올림하여 소숫점 이하 3자리)

from scipy.stats import binom

n = 21
p = 0.7
k = 16

prob = binom.cdf(k-1, n, p)
print(round(prob,3))
```

```python
# 5
# 적어도 19명이 치과를 찾았을 확률(반올림하여 소숫점 이하 3자리)

k = 19

prob = 1 - binom.cdf(k-1, n, p)
print(round(prob, 3))
```
