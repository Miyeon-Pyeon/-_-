```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/consum/Tetuan%20City%20power%20consumption.csv')
df.head()
```

```python
#51
# DateTime컬럼을 통해 각 월별로 몇개의 데이터가 있는지

df['DateTime'] = pd.to_datetime(df['DateTime'])

Ans = df['DateTime'].dt.month.value_counts().sort_index().to_frame()
print(Ans)
```

```python
#52
# 3월달의 각 시간대별 온도의 평균들 중 가장 낮은 시간대의 온도를  출력하라

target = df[df.DateTime.dt.month == 3]
Ans = target.groupby(target.DateTime.dt.hour)['Temperature'].mean().sort_values().values[0]
print(Ans)
```

```python
# 53
# 3월달의 각 시간대별 온도의 평균들 중 가장 높은 시간대의 온도를  출력하라

target = df[df.DateTime.dt.month == 3]
Ans = target.groupby(target.DateTime.dt.hour)['Temperature'].mean().sort_values().values[-1]
print(Ans)
```

```python
#54
# Zone 1 Power Consumption 컬럼의 value값의 크기가 Zone 2 Power Consumption 컬럼의 value값의 크기보다 큰 데이터들의 Humidity의 평균

Ans = df[df['Zone 1 Power Consumption'] > df['Zone 2  Power Consumption']].Humidity.mean()
print(Ans)
```

```python
#55
# 각 zone의 에너지 소비량의 상관관계를 구해서 데이터 프레임으로 표기

Ans = df.iloc[:,-3:].corr()
display(Ans)
```

```python
#56
# Temperature의 값이 10미만의 경우 A, 10이상 20미만의 경우 B,20이상 30미만의 경우 C, 그 외의 경우 D라고 할때 각 단계의 데이터 개수를 구하여라

def tem(x):
  if x < 10:
    return 'A'
  elif x <20:
    return 'B'
  elif x < 30:
    return 'C'
  else:
    return 'D'

df['tg'] = df.Temperature.map(tem)
Ans = df['tg'].value_counts()
display(Ans)
```

```python
#57
# 6월 데이터중 12시의 Temperature의 표준편차를 구하여라

Ans = df[(df.DateTime.dt.month == 6) & (df.DateTime.dt.hour == 12)].Temperature.std()
print(Ans)
```

```python
#58
# 6월 데이터중 12시의 Temperature의 분산을 구하여라

Ans = df[(df.DateTime.dt.month == 6) & (df.DateTime.dt.hour == 12)].Temperature.var()
print(Ans)
```

```python
#59
# Temperature의 평균이상의 Temperature의 값을 가지는 데이터를 Temperature를 기준으로 정렬 했을때 4번째 행의 Humidity 값은?

Ans = df[df['Temperature'] >= df['Temperature'].mean()].sort_values('Temperature').Humidity.iloc[3]
print(Ans)
```

```python
#60
# Temperature의 중간값 이상의 Temperature의 값을 가지는 데이터를Temperature를 기준으로 정렬 했을때 4번째 행의 Humidity 값은?

Ans = df[df['Temperature'] >= df['Temperature'].median()].sort_values('Temperature').Humidity.iloc[3]
print(Ans)
```
