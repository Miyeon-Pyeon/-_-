```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)
df.head()
```

```python
#1
# 인기동영상 제작횟수가 많은 채널 상위 10개 출력(채널ID의 빈도수 중 상위 10개를 뽑아 리스트로 출력)

# Ans = list(df.loc[df.channelId.isin(df.channelId.value_counts().head(10).index)].channelTitle.unique())
# print(Ans)

Ans = df['channelId'].value_counts().head(10).index
result = df[df.channelId.isin(Ans)].channelTitle.unique()
list(result)
```

```python
#2
#논란으로 인기동영상이 된 케이스 확인(dislike가 like보다 많은 동영상)

Ans = list(df.loc[df['dislikes']>df['likes']].channelTitle.unique())
print(Ans)
```

```python
#3
#채널명을 바꾼 케이스 확인(channelId는 고유값임)(중복되는 것을 제거한 뒤 빈도수를 계산하고 1이 아닌(채널명이 바뀐)채널의 수 구하기)

change = df[['channelTitle','channelId']].drop_duplicates().channelId.value_counts()
target = change[change>1]
print(len(target))
```

```python
#4
#일요일에 인기있었던 영상 중 가장 많은 영상의 종류

df ['trending_date2'] = pd.to_datetime(df['trending_date2'])

#Ans = df.loc[df['trending_date2'].dt.day_name() == 'Sunday'].categoryId.value_counts().index[0]
Ans = df[df['trending_date2'].dt.weekday == 6].categoryId.value_counts(1).index[0]     # weekday함수를 썼을 때 일요일은 6임

print(Ans)
```

```python
#5
# 각 요일별 인기 영상들의 categoryId는 각각 몇개씩인가


# pivot을 사용
# group = df.groupby([df['trending_date2'].dt.day_name(), 'categoryId'], as_index=False).size()
# Ans = group.pivot(index='categoryId', columns='trending_date2')
# display(Ans)

# pivot_table을 사용
# Ans = group.pivot_table(index='categoryId', columns='trending_date2')#, values='size', aggfunc='sum', fill_value=0)
# display(Ans)

# 결과는 같음. 결측치를 대체하거나 집계할 때는 pivot_table을 사용해야함

# crosstab 사용
df['name'] = df['trending_date2'].dt.day_name()
Ans = pd.crosstab(df['categoryId'], df['name'])
display(Ans)
```

```python
#6
# viewcount대비 댓글수가 가장 높은 영상(view_count값이 0인 경우 제외) => 댓글수를 조회수로 나눴을 때 가장 높은 영상

# target2 = df.loc[df.view_count != 0]
# t = target2.copy()
# t['ratio'] = (target2['comment_count']/target2['view_count']).dropna()
# result = t.sort_values(by='ratio', ascending=False).iloc[0].title
# print(result)

tg2 = df[df.view_count != 0].reset_index(drop = True)
tg2['ratio'] = tg2['comment_count'] / tg2['view_count']
result = tg2.sort_values('ratio')[['ratio','title']].title.values[-1]
print(result)
```

```python
#7
# viewcount대비 댓글수가 가장 낮은 영상(view_count, ratio값이 0인 경우 제외)

tg = tg2[(tg2['view_count'] != 0) & (tg2['ratio'] != 0)].reset_index(drop=True)
tg['ratio'] = tg['comment_count'] / tg['view_count']
result = tg.sort_values('ratio')[['ratio', 'title']].title.values[0]
print(result)
```

```python
#8
# 좋아요 대비 싫어요의 수가 가장 적은 영상

# like = df[(df['likes'] != 0) & (df['dislikes'] != 0)]
# like['ratio'] = df['dislikes'] / df['likes']
# result = like.sort_values('ratio').iloc[0]['title']
# print(result)

# target = df.loc[(df.likes != 0) & (df.dislikes != 0)]
# num = (target['dislikes'] / target['likes']).sort_values().index[0]
# Ans = df.iloc[num].title
# print(Ans)

dff = df[(df.likes != 0) & (df.dislikes != 0)].reset_index(drop=True)
dff['ratio'] = dff['dislikes'] / dff['likes']
result = dff.sort_values('ratio').title.values[0]
print(result)
```

```python
#9
# 가장 많은 트렌드 영상을 제작한 채널의 이름

dfc = df['channelTitle'].value_counts().index[0]
print(dfc)
```

```python
#10
#20회(20일)이상 인기동영상 리스트에 포함된 동영상의 수

# twt = (df['channelId'].value_counts() >= 20).sum()
# print(twt)

twt = df[['title','channelId']].value_counts().to_frame()
result = len(twt[twt['count']>=20])
print(result)
```
