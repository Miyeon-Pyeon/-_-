```python
import pandas as pd

df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/worldcup/worldcupgoals.csv')
df.head()
```

```python
#21
# 전체 기간의 각 나라별 골득점수 상위 5개 국가와 그 득점수를 데이터프레임

dff = df.groupby('Country').sum().sort_values('Goals',ascending=False).head(5).reset_index()
result = dff[['Country','Goals']]
display(result)
```

```python
#22
# 골 득점을 한 선수가 가장 많은 나라 상위 5개 국가와 그 선수 숫자를 데이터 프레임으로

dff = df.groupby('Country').size().sort_values(ascending=False).head(5)
display(dff)
```

```python
#23
# 년도 표기가 4자리 숫자로 안된 케이스가 몇건인가

df['yearLst'] = df.Years.str.split('-')

def checkFour(x):
  for value in x:
    if len(str(value)) != 4:
      return False
  return True

df['check'] = df['yearLst'].apply(checkFour)

result = len(df[df.check == False])
result
```

```python
#24
# 23에서 발생한 예외 케이스를 제외한 데이터프레임을 df2라고 정의하고 데이터의 행의 숫자를 출력하라 (아래 문제부터는 df2로 풀이하겠습니다)

df2 = df[df['check'] == True].reset_index(drop=True)
print(len(df2))
```

```python
#25
# 월드컵 출전횟수를 나타내는 ‘LenCup’ 컬럼을 추가하고 4회 출전한 선수의 숫자를 구하여라

df2['LenCup'] = df2['yearLst'].str.len()
result = df2['LenCup'].value_counts()[4]
print(result)
```

```python
#26
# Yugoslavia 국가의 월드컵 출전횟수가 2회인 선수들의 숫자를 구하여라

yugo = len(df2[(df2.LenCup == 2) &(df2['Country'] == 'Yugoslavia')])
print(yugo)
```

```python
#27
# 2002년도에 출전한 전체 선수는 몇명인가

two = len(df2[df2['Years'].str.contains('2002')])
print(two)
```

```python
#28
# 이름에 'carlos'가 들어가는 선수는 몇명인가

carlos = len(df2[df2['Player'].str.lower().str.contains('carlos')])
print(carlos)
```

```python
#29
# 월드컵 출전 횟수가 1회뿐인 선수들 중에서 가장 많은 득점을 올렸던 선수는?

one = df2[df2.LenCup == 1].sort_values('Goals',ascending = False).Player.values[0]
print(one)
```

```python
#30
# 월드컵 출전 횟수가 1회뿐인 선수들이 가장 많은 국가는?

most = df2[df2.LenCup == 1].Country.value_counts().index[0]
print(most)
```
