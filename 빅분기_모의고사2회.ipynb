```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
df.head()
```

```python
# 1
# 성별이 Male인 환자들의 age의 평균값은 ?

df['age'] = df['age'].str.replace('*','').astype('int')
Ans = df[df['gender'] == 'Male'].age.mean()
print(Ans)
```

```python
# 2
# bmi컬럼의 결측치를 bmi컬럼의 결측치를 제외한 나머지 값들의 중앙값으로 채웠을 경우 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

Ans = df['bmi'].fillna(df['bmi'].median()).mean()
print(round(Ans, 3))
```

```python
# 3
# bmi컬럼의 각 결측치들을 직전의 행의 bmi값으로 채웠을 경우 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

Ans = df['bmi'].fillna(method = 'ffill').mean()
print(round(Ans, 3))
```

```python
# 4
# bmi컬럼의 각 결측치들을 결측치를 가진 환자 나이대(10단위)의 평균 bmi 값으로 대체한 후 대체된 bmi 컬럼의 평균을 소숫점 이하 3자리 까지 구하여라

df['age'] = df['age'].replace('*','').astype('int')

mean = df[df.bmi.notnull()].groupby(df.age // 10 *10).bmi.mean()
dic = {x:y for x,y in mean.items()}

idx = df.loc[df.bmi.isnull(), ['age','bmi']].index

df.loc[df.bmi.isnull(),'bmi'] = (df[df.bmi.isnull()].age//10*10).map(lambda x: dic[x])

result = df.bmi.mean()
print(round(result,3))
```

```python
# 5
#avg_glucose_level 컬럼의 값이 200이상인 데이터를 모두 199로 변경하고 stroke값이 1인 데이터의 avg_glucose_level값의 평균을 소수점이하 3자리 까지 구하여라

df.loc[df.avg_glucose_level >= 200, 'avg_glucose_level'] = 199
Ans = df[df['stroke'] == 1].avg_glucose_level.mean()
print(round(Ans,3))
```

```python
# 7
# Type 1 컬럼의 종류에 따른 Total 컬럼의 평균값을 내림차순 정렬했을때 상위 3번째 Type 1은 무엇인가?

dd = df.groupby('Type 1').Total.mean().sort_values(ascending = False).index[2]
print(dd)
```

```python
# 8
# 결측치가 존재하는 행을 모두 지운 후 처음부터 순서대로 60% 데이터를 추출하여 Defense컬럼의 1분위수를 구하여라

ans = df.dropna()[:int(len(df.dropna())*0.6)].Defense.quantile(0.25)
print(ans)
```

```python
# 9
# Type 1 컬럼의 속성이 Fire인 포켓몬들의 Attack의 평균이상인 Water속성의 포켓몬 수를 구하여라

ans = (df[df['Type 1'] == 'Water'].Attack >= (df[df['Type 1'] == 'Fire'].Attack.mean())).sum()
print(ans)
```

```python
# 10
# 각 세대 중(Generation 컬럼)의 Speed와 Defense 컬럼의 차이(절댓값)이 가장 큰 세대는?

Ans = abs(df.groupby(['Generation'])[['Speed','Defense']].mean().T.diff().T).sort_values('Defense').index[-1]
print(Ans)
```

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/test.csv')

display(train.head())
display(test.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import LabelEncoder
```

```python
print(train.info()) # 결측치: bmi / 라벨링: gender, age, ever_married, work_type, Residence_type, smoking_status
print(test.info())
```

```python
# 결측치

train.fillna(train['bmi'].mean(), inplace=True)

print(train.info())
```

```python
# 인코딩

le = LabelEncoder()

train['gender'] = le.fit_transform(train['gender'])
test['gender'] = le.fit_transform(test['gender'])

train['age'] = le.fit_transform(train['age'])
test['age'] = le.fit_transform(test['age'])

train['ever_married'] = le.fit_transform(train['ever_married'])
test['ever_married'] = le.fit_transform(test['ever_married'])

train['work_type'] = le.fit_transform(train['work_type'])
test['work_type'] = le.fit_transform(test['work_type'])

train['Residence_type'] = le.fit_transform(train['Residence_type'])
test['Residence_type'] = le.fit_transform(test['Residence_type'])

train['smoking_status'] = le.fit_transform(train['smoking_status'])
test['smoking_status'] = le.fit_transform(test['smoking_status'])
```

```python
print(train.info()) # 결측치: bmi / 라벨링: hypertension, heart_disease, smoking_status
print(test.info())
```

```python
y = train['stroke']
X = train.drop(columns = ['stroke', 'id'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size =0.3, random_state=42)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
pred = rf.predict_proba(x_test)
```

```python
print('roc : ', roc_auc_score(y_test, pred[:,1]))
```
