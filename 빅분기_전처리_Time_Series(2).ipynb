```python
#67

#나의 오답
# def cyear(x):
#   if x >= 2061:
#     return x-100

# df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(cyear)

# Ans = df.head()



def fix_century(x):
  import datetime

  year = x.year - 100 if x.year >= 2061 else x.year
  return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)

df.head()
```

```python
#68
# 년도별 각 칼럼의 평균값을 구하라

Ans = df.groupby(df.Yr_Mo_Dy.dt.year).mean()
Ans.head()
```

```python
df = pd.read_csv('http://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
df.head()
```

```python
#76

def change_date(x):
  import datetime
  hour = x.split(':')[1]
  date = x.split(':')[0]

  if hour == '24':
    hour = '00:00:00'

    FinalDate = pd.to_datetime(date + " " + hour) + datetime.timedelta(days=1)

  else:
    hour = hour + ':00:00'
    FinalDate = pd.to_datetime(date + " " + hour)
  return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

Ans = df
Ans.head()
```

```python
#77

df['dayName'] = df['(년-월-일:시)'].dt.day_name()
Ans = df['dayName'].to_frame()
Ans.head()
```

```python
#78

Ans = df.groupby(['dayName','PM10등급'], as_index= False).size()
Ans.head()

Ans2 = Ans.pivot(index='dayName', columns='PM10등급', values='size').fillna(0)
Ans2
```
