```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

display(train.head(2))
test.head(2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

import numpy as np
```

```python
print(train.info())
print(test.info())
# 결측치는 없음
# 라벨링: 업종명, 이용자구분, 성별
```

```python
le = LabelEncoder()

train['업종명'] = le.fit_transform(train['업종명'])
train['이용자구분'] = le.fit_transform(train['이용자구분'])
train['성별'] = le.fit_transform(train['성별'])

test['업종명'] = le.fit_transform(test['업종명'])
test['이용자구분'] = le.fit_transform(test['이용자구분'])
test['성별'] = le.fit_transform(test['성별'])

print(train.info())
print(test.info())
```

```python
y = train['이용금액']
X = train.drop(columns = ['이용금액','ID'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,test_size = 0.4, random_state=42)
```

```python
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
pred = rf.predict(x_test)
```

```python
print('rmse : ', np.sqrt(mean_squared_error(y_test, pred)))
```

```python
test_X = test.drop(columns=['ID'])
```

```python
pred2 = rf.predict(test_X)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['price'] = pred2
submission.head()
```
