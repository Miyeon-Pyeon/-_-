## 5회 기출

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

display(train.head(2))
test.head(2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
print(train.info())   # 수치형: model, transmission, fuelType
print(test.info())
```

```python
le = LabelEncoder()

train['model'] = le.fit_transform(train['model'])
test['model'] = le.fit_transform(test['model'])

train['transmission'] = le.fit_transform(train['transmission'])
test['transmission'] = le.fit_transform(test['transmission'])

train['fuelType'] = le.fit_transform(train['fuelType'])
test['fuelType'] = le.fit_transform(test['fuelType'])
```

```python
print(train.info())   # 수치형: model, transmission, fuelType
print(test.info())
```

```python
y = train['price']
X = train.drop(columns=['price','ID'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
pred = rf.predict(x_test)
```

```python
test_x = test.drop(columns = ['ID'])
```

```python
print('rmse : ', np.sqrt(mean_squared_error(y_test, pred)))
```

```python
pred2 = rf.predict(test_x)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['price'] = pred2
submission.head()
#submission.to_csv('result.csv', index=False)
```

## 4회 기출

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_test.csv')

display(train.head(2))
test.head(2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.preprocessing import LabelEncoder
```

```python
# print(train.info())   # 결측치 : Ever_Married, Graduated, Profession, Work_Experience, Family_Size, Var_1
# print(test.info())    # 수치화 : Gender, Ever_Married, Graduated, Profession, Spending_Score, Var_1
```

```python
# 결측치는 아무값으로 대체함. => 정확도 낮음

train.Ever_Married = train['Ever_Married'].fillna(0)
test.Ever_Married = test['Ever_Married'].fillna(0)

train.Graduated = train['Graduated'].fillna('Yes')
test.Graduated = test['Graduated'].fillna('Yes')

train.Profession = train['Profession'].fillna('Artist')
test.Profession = test['Profession'].fillna('Artist')

train.Work_Experience = train['Work_Experience'].fillna(train['Work_Experience'].mean())
test.Work_Experience = test['Work_Experience'].fillna(train['Work_Experience'].mean())

train.Family_Size = train['Family_Size'].fillna(train['Family_Size'].mean())
test.Family_Size = test['Family_Size'].fillna(train['Family_Size'].mean())

train.Var_1 = train['Var_1'].fillna('Cat_6')
test.Var_1 = test['Var_1'].fillna('Cat_6')
```

```python
le = LabelEncoder()

train['Gender'] = le.fit_transform(train['Gender'])
test['Gender'] = le.fit_transform(test['Gender'])

train['Ever_Married'] = le.fit_transform(train['Ever_Married'])
test['Ever_Married'] = le.fit_transform(test['Ever_Married'])

train['Graduated'] = le.fit_transform(train['Graduated'])
test['Graduated'] = le.fit_transform(test['Graduated'])

train['Profession'] = le.fit_transform(train['Profession'])
test['Profession'] = le.fit_transform(test['Profession'])

train['Spending_Score'] = le.fit_transform(train['Spending_Score'])
test['Spending_Score'] = le.fit_transform(test['Spending_Score'])

train['Var_1'] = le.fit_transform(train['Var_1'])
test['Var_1'] = le.fit_transform(test['Var_1'])
```

```python
y = train['Segmentation']
X = train.drop(columns = ['Segmentation','ID'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
pred = rf.predict(x_test)
```

```python
print('f1_score: ', f1_score(y_test, pred, average='macro'))
```

```python
test_x = test.drop(columns = ['ID'])
```

```python
pred2 = rf.predict(test_x)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['Segmentation'] = pred2
submission.head()
#submission.to_csv('result.csv', index=False)
```
