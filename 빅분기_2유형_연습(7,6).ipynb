## 기출 7회

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

display(train.head(2))
test.head(2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
print(train.info()) # 수치화 : 업종명, 이용자구분, 성별
print(test.info())
```

```python
le = LabelEncoder()

train['업종명'] = le.fit_transform(train['업종명'])
test['업종명'] =le.fit_transform(test['업종명'])

train['이용자구분'] = le.fit_transform(train['이용자구분'])
test['이용자구분'] =le.fit_transform(test['이용자구분'])

train['성별'] = le.fit_transform(train['성별'])
test['성별'] =le.fit_transform(test['성별'])
```

```python
print(train.info()) # 수치화 완료
print(test.info())
```

```python
y = train['이용금액']
X = train.drop(columns = ['이용금액', 'ID'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
pred = rf.predict(x_test)
```

```python
print('rmse :', np.sqrt(mean_squared_error(y_test, pred)))
```

```python
test_x = test.drop(columns = ['ID'])
```

```python
pred2 = rf.predict(test_x)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['price'] = pred2
submission.head()
#submission.to_csv('result.csv', index=False)
```

## 기출 6회

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_test.csv')

display(train.head(2))
test.head(2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.preprocessing import LabelEncoder
```

```python
print(train.info())
print(test.info())
```

```python
le = LabelEncoder()

train['Checkup'] = le.fit_transform(train['Checkup'])
test['Checkup'] = le.fit_transform(test['Checkup'])

train['Exercise'] = le.fit_transform(train['Exercise'])
test['Exercise'] = le.fit_transform(test['Exercise'])

train['Heart_Disease'] = le.fit_transform(train['Heart_Disease'])
test['Heart_Disease'] = le.fit_transform(test['Heart_Disease'])

train['Skin_Cancer'] = le.fit_transform(train['Skin_Cancer'])
test['Skin_Cancer'] = le.fit_transform(test['Skin_Cancer'])

train['Other_Cancer'] = le.fit_transform(train['Other_Cancer'])
test['Other_Cancer'] = le.fit_transform(test['Other_Cancer'])

train['Depression'] = le.fit_transform(train['Depression'])
test['Depression'] = le.fit_transform(test['Depression'])

train['Diabetes'] = le.fit_transform(train['Diabetes'])
test['Diabetes'] = le.fit_transform(test['Diabetes'])

train['Arthritis'] = le.fit_transform(train['Arthritis'])
test['Arthritis'] = le.fit_transform(test['Arthritis'])

train['Sex'] = le.fit_transform(train['Sex'])
test['Sex'] = le.fit_transform(test['Sex'])

train['Age_Category'] = le.fit_transform(train['Age_Category'])
test['Age_Category'] = le.fit_transform(test['Age_Category'])

train['Smoking_History'] = le.fit_transform(train['Smoking_History'])
test['Smoking_History'] = le.fit_transform(test['Smoking_History'])
```

```python
print(train.info())
print(test.info())
```

```python
y = train['General_Health']
X = train.drop(columns = ['General_Health', 'ID'])
```

```python
x_test, x_train, y_test, y_train = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
pred = rf.predict(x_test)
```

```python
print('f1 score: ', f1_score(y_test, pred, average='macro'))
```

```python
test_x = test.drop(columns=['ID'])
```

```python
pred2 = rf.predict(test_x)
```

```python
submission = pd.DataFrame()
submission['ID'] = test['ID']
submission['General_Health'] = pred2
submission.head()
# submission.to_csv('result.csv', index=False)
```
