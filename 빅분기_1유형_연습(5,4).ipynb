## 5회 기출

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_1_.csv')
df.head(5)
```

```python
# 1
# 20L가격과 5L가격이 모두 0원이 아닌 데이터만 필터를 한 후, 각 row별로 20L가격과 5L가격의 차이를 ‘차이가격’ 이라 부른다고 하자. 시도명 별 차이가격의 평균가격을 비교할때 그 값이 가장 큰 금액을 반올림하여 소숫점 이하 1자리까지 구하여라

noze = df[(df['20L가격']!=0) & (df['5L가격']!=0)].dropna()
noze['차이가격'] = noze['20L가격'] - noze['5L가격']
ans = noze.groupby('시도명').차이가격.mean().max()
print(round(ans,1))
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_2_.csv')
df.head(5)
```

```python
# 2
# BMI는 몸무게(kg) / (키(M) * 키(M)) 로 정의 된다. 초고도 비만은 BMI 25이상 , 고도 비반은 BMI 25미만 - 23이상 , 정상은 23미만 - 18.5이상 저체중은 18.5미만으로 정의 된다. 주어진 데이터에서 초고도비만 인원 + 저체중 인원 의 숫자는?

df['BMI'] = df['weight(kg)'] / ((df['height(cm)']/100) * (df['height(cm)']/100))

def fat(x):
  if x >= 25:
    return '초고도비만'
  elif x >= 23:
    return '고도비만'
  elif x >= 18.5:
    return '정상'
  else:
    return '저체중'

df['bmi_category'] = df['BMI'].map(fat)
Ans = df[df['bmi_category'].isin(['초고도비만','저체중'])].shape[0]
print(Ans)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_3.csv')
df.head(5)
```

```python
# 3
# 순유입인원은 초중고 도내,도외 전입인원에서 초중고 도내, 도외 전출인원을 뺀값이다. 각년도별로 가장 큰 순유입인원을 가진 지역구의 순유입인원을 구하고 전체 기간의 해당 순유입인원들의 합을 구하여라

in_ = df['고등학교_전입_도내'] + df['고등학교_전입_도외'] + df['중학교_전입_도내'] + df['중학교_전입_도외'] + df['초등학교_전입_도내'] + df['초등학교_전입_도외']
out_ = df['고등학교_전출_도내'] + df['고등학교_전출_도외'] + df['중학교_전출_도내'] + df['중학교_전출_도외'] + df['초등학교_전출_도내'] + df['초등학교_전출_도외']

df['순유입인원'] = in_ - out_

ans = (df.groupby('년도')[['순유입인원', '지역']].max().sort_values(by='순유입인원')).sum()
print(ans.values[0])
```

## 4회 기출

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p1_1.csv')
df.head(5)
```

```python
# 1
# Temperature컬럼에서 숫자가 아닌 문자들을 제거후 숫자 타입으로 바꾸고 3분위수에서 1분위수의 차이를 소숫점 이하 2자리까지 구하여라

df.loc[:,'Temperature'] = df['Temperature'].replace('*',' ', regex=False).astype('float')
Ans = df['Temperature'].quantile(0.75) - df['Temperature'].quantile(0.25)
print(round(Ans, 2))
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p1_2.csv')
df.head(5)
```

```python
# 2
# Likes를 Comments로 나눈 비율이 20이상이면서 Keyword값이 minecraft인 영상들의 Views값의 평균을 정수로 구하여라

ans = int(df[((df['Likes'] / df['Comments']) >= 20) & (df['Keyword'] == 'minecraft')].Views.mean())
print(ans)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p1_3.csv')
df.head(5)
```

```python
# 3
# date_added가 2018년 1월 이면서 country가 United Kingdom 단독 제작인 데이터의 갯수

df['date_added_dt'] = pd.to_datetime(df['date_added'], errors='coerce')
ans = df[(df['date_added_dt'].dt.strftime('%Y-%m') == '2018-01') & (df['country'] == 'United Kingdom')].shape[0]
print(ans)
```
