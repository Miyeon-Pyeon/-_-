## 7회 기출

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3_1.csv')
df.head()
```

```python
# 1
# 선형관계 가장 큰 변수 찾아 상관계수를 구하여라

ans = df.corr()['Target'].value_counts().sort_index().index[-2]
print(ans)
```

```python
# 2
# Target 변수를 종속변수로 하여 다중선형회귀모델링을 진행했을 때 v2 컬럼의 회귀 계수는?

import numpy as np
import statsmodels.api as sm

y = df['Target']
X = df.drop(columns = ['Target'])
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.params['v2'])
```

```python
# 3
# 회귀 계수들이 가지는 p값들 중 최대 값은?

ans = model.pvalues.max()
print(ans)
```

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3_t.csv')
df.head()
```

```python
# 4
# train 데이터로 target을 종속변수로 로지스틱 회귀를 진행할 때 age 컬럼의 오즈비를 구하여라

import statsmodels.api as sm

train = df.iloc[:200].reset_index(drop=True)
test = df.iloc[200:].reset_index(drop=True)

y = train['target']
X = train.drop(columns = ['target'])



model = sm.Logit(y,X).fit()

odds_ratio = np.exp(model.params['age'])
print(odds_ratio)
```

```python
# 5
# train으로 로지스틱 회귀 진행했을 경우 잔차 이탈도 (residual deviance)를 계산하라

model2 = sm.GLM(y, X, family=sm.families.Binomial()).fit()

residual_deviance = model2.deviance
print(residual_deviance)
```

```python
# 6
# train으로 로지스틱 회귀 진행했을 경우 로짓 우도값을 도출하라

X = train.drop('target', axis=1)
X = sm.add_constant(X)
y = train['target']

model = sm.Logit(y, X).fit()

print(model.llf)
```

```python
# 7
# test 데이터의 독립변수로 target 예측 후 오류율을 구하여라

pred = (model2.predict(test.drop(columns=['target'])) >0.5).astype('int')

from sklearn.metrics import accuracy_score

error_rate = 1 - accuracy_score(test['target'], pred)
print(error_rate)
```

## 6회 기출

```python
# A 도시의 남성 600명과 여성 550명이 있다. 남성들 중 흡연자 비율은 0.2이며 여성들 중 흡연자 비율은 0.26이다.
# 남성과 여성 간에 흡연 여부에 따른 인구 비율이 다른지 확인하고 싶다. 유의 수준 0.05하 귀무가설에 대해 기각 / 채택 여부와 p-value값을 각각 출력하라
```

```python
# 1

import numpy as np

male = 600
female = 550

m_smoking_ratio = 0.2
f_smoking_ratio = 0.26

smoking_male = male * m_smoking_ratio
non_smoking_male = male - smoking_male

smoking_female = female * f_smoking_ratio
non_smoking_female = female - smoking_female

from scipy.stats import chi2_contingency

data = np.array([[smoking_male, non_smoking_male], [smoking_female, non_smoking_female]])

chi2, p_val, _, _ = chi2_contingency(data)

print(round(p_val, 3))

# 0.05보다 작으므로 기각
```

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/28/p7.csv')
df.head()
```

```python
# 2
# age와 Cholesterol을 가지고 weight를 예측하는 선형 회귀 모델을 만들려고한다. age의 회귀 계수를 구하여라

import statsmodels.api as sm

X = sm.add_constant(df[['age','Cholesterol']])
model = sm.OLS(df['weight'], X)
Ans = model.fit()

print(Ans.params['age'])
```

```python
# 3
# age가 고정일 때 Cholesterol와 weight가 선형관계에 있다는 가설을 유의수준 0.05하에 검정하라

H = Ans.pvalues['Cholesterol']

if H < 0.05:
  print('선형관계에 있다.')
else:
  print('선형관계에 없다.')
```

```python
# 4
# age가 55, Cholesterol가 72.6일때 위 모델을 기반으로 weight값을 예측하라

pred = Ans.predict([1,55,72.6])
print(pred[0])
```
