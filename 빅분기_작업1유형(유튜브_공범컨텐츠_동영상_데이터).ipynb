```python
import pandas as pd

ch = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
vd = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
```

```python
display(ch.head())
display(vd.head())
```

```python
#11
# 'ct'칼럼을 시간으로 인식할 수 있게 datatype을 변경하고 video데이터의 videoname의 빈도수 확인

vd['ct'] = pd.to_datetime(vd['ct'])
Ans = vd.videoname.value_counts()
print(Ans)
```

```python
#12
# 각 video의 가장 최신화 된 날짜의 viewcount값

last = vd.sort_values(['videoname','ct']).reset_index(drop=True).drop_duplicates(['videoname','ct'], keep='last')[['viewcnt','videoname','ct']].reset_index(drop=True)
display(last)
```

```python
#13
# channel데이터 중 2021-10-03 이후 각 채널의 처음 기록됐던 구독자 수

ch['ct'] = pd.to_datetime(ch['ct'])
sub = ch[ch.ct >= pd.to_datetime('2021-10-03')].sort_values(['ct','channelname']).drop_duplicates('channelname')
Ans = sub[['channelname','subcnt']].reset_index(drop=True)
print(Ans)
```

```python
#14
# 각 채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00까지의 구독자수 증가량

start = ch.loc[ch.ct.dt.strftime('%Y-%m-%d %H') =='2021-10-03 03']
end = ch.loc[ch.ct.dt.strftime('%Y-%m-%d %H') =='2021-11-01 15']

start_df = start[['channelname','subcnt']].reset_index(drop=True)
end_df = end[['channelname','subcnt']].reset_index(drop=True)

start_df.columns = ['channelname','start_df']
end_df.columns = ['channelname','end_df']

sub = pd.merge(start_df, end_df)
sub['del'] = sub['end_df'] - sub['start_df']
result = sub[['channelname','del']]
display(result)
```

```python
#15
# 공범EP1의 비디오정보 데이터 중 수집 간격이 5분 이하, 20분 이상인 데이터 구간의 시각을 출력

import datetime

ep_one = vd.loc[vd.videoname.str.contains('1')].sort_values('ct').reset_index(drop = True)

ep_one[(ep_one.ct.diff(1) >= datetime.timedelta(minutes=20)) | (ep_one.ct.diff(1) >= datetime.timedelta(minutes=5))]

Ans = ep_one[ep_one.index.isin([720,721,722,723,1635,1636,1637])]

display(Ans)
```

```python
#16
# 각 에피소드의 시작 날짜를 에피소드의 이름과 묶어 데이터 프레임으로 출력

dff = vd.sort_values(['ct','videoname']).drop_duplicates('videoname')[['ct','videoname']]
dff['date'] = dff.ct.dt.date
Ans = dff[['date','videoname']]
display(Ans)
```

```python
#17
# 공개된 날의 21시의 viewcnt, ct, videoname으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬

vd['time'] = vd.ct.dt.hour

Ans = vd.loc[vd['time'] == 21].sort_values(['videoname','ct']).drop_duplicates('videoname').sort_values('viewcnt', ascending = False)[['videoname','viewcnt','ct']].reset_index(drop=True)

display(Ans)
```

```python
#18
# video정보의 가장 최근 데이터들에서 각 에피의 싫어요/좋아요 비율을 ratio컬럼으로 만들고 videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬

vd['ratio'] = vd.dislikecnt / vd.likecnt

Ans = vd[['videoname','ratio']].drop_duplicates('videoname',keep = 'last').sort_values('ratio',ascending = True).reset_index(drop=True)

display(Ans)
```

```python
#19
# 2021-11-01 00:00:00 ~ 15:00:00 까지 각 에프소드별 viewcnt의 증가량을 데이터 프레임으로

start = pd.to_datetime('2021-11-01 00:00:00')
end = pd.to_datetime('2021-11-01 15:00:00')

tar = vd.loc[(vd['ct'] >= start) & (vd['ct'] <= end)].reset_index(drop = True)

def check(x):
  result = max(x) - min(x)
  return result

Ans = tar[['videoname','viewcnt']].groupby('videoname').agg(check)
Ans
```

```python
#20
# vd데이터 중 중복되는 데이터가 존재한다. 중복되는 각 데이터의 시간대와 videoname을 구하라

Ans = vd[vd.index.isin(set(vd.index) - set(vd.drop_duplicates().index))]
result = Ans[['videoname','ct']]
display(result)
```
