```python
import pandas as pd
```

```python
#64

df = pd.read_csv('http://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv')
df.info()
```

```python
#65

df.Yr_Mo_Dy = pd.to_datetime(df.Yr_Mo_Dy)
Ans = df.Yr_Mo_Dy
Ans.head()
```

```python
#66

Ans = df.Yr_Mo_Dy.dt.year.unique()
Ans
```

```python
df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].fillna(0)
```

```python
#67

#나의 오답
# def cyear(x):
#   if x >= 2061:
#     return x-100

# df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(cyear)

# Ans = df.head()



def fix_century(x):
  import datetime

  year = x.year - 100 if x.year >= 2061 else x.year
  return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)

df.head()
```
